```python
def con(value):

    if value == 1:
        return "Yes"
    else:
        return "No"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(con)
```

```python
ax = sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0]) 
plt.title("Count of churned customers")
plt.show()
```

```python
gb = df.groupby('Churn').agg({'Churn':'count'})

plt.pie(gb['Churn'], labels = gb.index ,autopct ="%1.2f%%")
plt.title("Percentage of churned customers")
plt.show()
```

```python

#Create a crosstab of counts
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert counts to percentages
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = ct_percent.plot(kind='bar', stacked=True, colormap='Set2')

# Add percentage labels to each segment
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

# Add title and labels
plt.title("SeniorCitizen Analysis (Churn %)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python


# Load your dataset
# Example: df = pd.read_csv('your_data.csv')

columns_to_plot = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
    'PaperlessBilling', 'PaymentMethod'
]

# Set up subplot grid
num_plots = len(columns_to_plot)
cols = 4
rows = (num_plots + cols - 1) // cols
fig, axes = plt.subplots(rows, cols, figsize=(20, 5 * rows))
axes = axes.flatten()

# Generate stacked bar charts with percentage labels
for i, column in enumerate(columns_to_plot):
    ax = axes[i]
    ct = pd.crosstab(df[column], df['Churn'])
    ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100
    ct_percent.plot(kind='bar', stacked=True, colormap='Set2', ax=ax)
    
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f%%', label_type='center')
    
    ax.set_title(f"{column} Analysis (Churn %)")
    ax.set_xlabel(column)
    ax.set_ylabel("Percentage")
    ax.legend(title="Churn")
    ax.tick_params(axis='x', rotation=45)

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```



 Demographic Factors
- Gender: Churn rates are slightly higher for females (28.9%) than males (26.2%), but the difference is marginal.
- Senior Citizens: A significant churn gap exists—41.7% of senior citizens churned vs. 23.6% of non-seniors.
- Partner Status: Customers without partners churn more (33%) than those with partners (19.7%).
- Dependents: Customers without dependents churn at 31.3%, while those with dependents churn at only 15.5%.

---

  Service Usage
- Phone Service: Churn is similar whether customers have phone service or not (~25%).
- Multiple Lines: Slightly higher churn for customers with multiple lines (28.6%) vs. those without (25.2%).

---

  Internet & Security Services
- Internet Service:
  - Fiber optic users churn the most (41.9%).
  - DSL users churn less (19.6%).
  - No internet users churn the least (8.8%).
- Online Security: Customers without security churn at 42.9%, while those with it churn at only 14.6%.
- Online Backup: Churn is 39.9% without backup vs. 21.5% with backup.
- Device Protection: Slight churn reduction with protection (22.5%) vs. without (27.0%).
- Tech Support: Huge churn gap—41.9% without support vs. 15.2% with support.

---

  Streaming Services
- Streaming TV and Movies show similar churn rates:
  - With service: ~30%
  - Without service: ~33%

---

  Contract Type
- Month-to-month contracts have the highest churn (43%).
- One-year contracts: 11.3% churn.
- Two-year contracts: only 3% churn.
>  Longer contracts strongly correlate with lower churn.

---

  Billing & Payment
- Paperless Billing: Higher churn (33.6%) vs. non-paperless (16.37%).
- Payment Method:
  - Electronic check users churn the most (44.45%).
  - Mailed check users churn less (19.45%).
  - Bank transfer and credit card (automatic) users have similar churn (~22.9%).

---

  Key Takeaways
- Senior citizens, month-to-month contract holders, and electronic check payers are most likely to churn.
- Support services (tech support, online security, backup) significantly reduce churn.
- Long-term contracts and dependents are strong retention indicators.
